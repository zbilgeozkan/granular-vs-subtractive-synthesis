```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import os

#  Fonksiyon: WAV dosyasn oku ve dalga formunu iz
def plot_waveform(file_path):
    # 1. Dosyay oku
    fs, data = wavfile.read(file_path)  # fs: rnekleme hz, data: ses verisi

    # 2. Stereo ise mono'ya evir
    if data.ndim > 1:
        data = data[:, 0]

    # 3. Normalize et
    data = data.astype(np.float32)
    data /= np.max(np.abs(data))

    # 4. Zaman ekseni (saniye cinsinden)
    duration = len(data) / fs
    time = np.linspace(0, duration, num=len(data))

    # 5. Dalga formunu iz
    plt.figure(figsize=(12, 4))
    plt.plot(time, data)
    plt.title(f"{os.path.basename(file_path)}  Dalga Formu ({duration:.2f} saniye)")
    plt.xlabel("Zaman (s)")
    plt.ylabel("Genlik")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

#  rnek kullanm
plot_waveform("sample.wav")  # Kendi dosya yolunu buraya gir
plot_waveform("subtractive_output.wav")
plot_waveform("granular_output.wav")
```
